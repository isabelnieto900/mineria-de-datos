```python

import numpy as np
import matplotlib.pyplot as plt
import time
```

# NUMPY VS PYTHON

```python
dim = []
times_for = []
times_np = []

for n in range(2,101):
    dim.append(n)
    matrix_a = np.random.randint(0, 10, size=(n, n))
    matrix_b = np.random.randint(0, 10, size=(n, n))
    # Product of two matrix by two cicles for
    start = time.time()
    for i in range(n):
        for j in range(n):
            matrix_a[i,j] + matrix_b[i,j]
    times_for.append(time.time()-start)

    # Product of two matrix by numpy
    start = time.time()
    matrix_c = np.dot(matrix_a,matrix_b)
    times_np.append(time.time()-start)
"""
print("Number of dimension", dim)
print("Times for:", times_for)
print("Times numpy:", times_np)
"""
plt.plot(dim, times_for, label='For', color='palevioletred')
plt.plot(dim, times_np, label='Numpy', color='rebeccapurple')
plt.xlabel('Dimensin de la matriz (n)')
plt.ylabel('Tiempo de ejecucin (segundos)')
plt.text(max(dim)*0.8, min(times_for)*1.1, 'Isabel Nieto', fontsize=10, color='black')
plt.grid()
plt.legend()
plt.show()
```
